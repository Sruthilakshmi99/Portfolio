```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import numpy as np
```

```python
stroke_data = pd.read_csv("C:\\Users\\vrsru\\Downloads\\healthcare-dataset-stroke-data.csv")
loan_data = pd.read_csv("C:\\Users\\vrsru\\Downloads\\loan_data.csv")
```

```python
#referenced from lab 1 and lab2 
def correlation_heatmap(data, num_features=7, dataset_name="Dataset"):
    numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()
    if len(numeric_columns) < num_features:
        num_features = len(numeric_columns)
    
    features_mean = numeric_columns[:num_features]

    plt.figure(figsize=(10, 8))
    sns.heatmap(data[features_mean].corr(), annot=True, square=True, cmap='coolwarm')
    plt.xticks(rotation=25, fontsize=10)
    plt.yticks(rotation=0, fontsize=10)
    plt.title(f"Correlation Heatmap - {dataset_name}")
    plt.show()
```

```python
def scatter_matrix_plot(data, num_features=7, dataset_name="Dataset"):
    numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()
    if len(numeric_columns) < num_features:
        num_features = len(numeric_columns)

    features_mean = numeric_columns[:num_features]

    color_dic = {0: 'red', 1: 'blue'}  # Example color mapping for binary targets
    target_column = 'target' if 'target' in data.columns else None

    if target_column and target_column in data.columns:
        target_list = data[target_column].tolist()
        colors = list(map(lambda x: color_dic.get(x, 'gray'), target_list))
    else:
        colors = 'gray'

    scatter_matrix(data[features_mean], c=colors, alpha=0.4, figsize=(10, 10))
    plt.suptitle(f"Scatter Matrix - {dataset_name}")
    plt.show()
```

```python
def histogram_plot(data, num_features=7, dataset_name="Dataset"):
    numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()
    if len(numeric_columns) < num_features:
        num_features = len(numeric_columns)

    features_mean = numeric_columns[:num_features]

    data[features_mean].hist(bins=15, figsize=(10, 10), alpha=0.7, color='blue', edgecolor='black')
    plt.suptitle(f"Histogram - {dataset_name}")
    plt.show()
```

```python
def boxplot_visualization(data, num_features=7, dataset_name="Dataset"):
    numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()
    if len(numeric_columns) < num_features:
        num_features = len(numeric_columns)

    features_mean = numeric_columns[:num_features]

    data[features_mean].plot(kind='box', subplots=True, layout=(3, 3), sharex=False, sharey=False, figsize=(10, 10))
    plt.suptitle(f"Boxplot - {dataset_name}")
    plt.show()
```

```python
correlation_heatmap(stroke_data, dataset_name="Stroke Prediction Dataset")
```

```python
scatter_matrix_plot(stroke_data, dataset_name="Stroke Prediction Dataset")
```

```python
histogram_plot(stroke_data, dataset_name="Stroke Prediction Dataset")
```

```python
boxplot_visualization(stroke_data, dataset_name="Stroke Prediction Dataset")
```

```python
correlation_heatmap(loan_data, dataset_name="Loan Data Dataset")
```

```python
scatter_matrix_plot(loan_data, dataset_name="Loan Data Dataset")
```

```python
histogram_plot(loan_data, dataset_name="Loan Data Dataset")
```

```python
boxplot_visualization(loan_data, dataset_name="Loan Data Dataset")
```
